```python
# We need to import the programs we want to use
# Using "as" allows us to reference the program without writing the full name every time
# Imports separated by a period "." are importing just part of the parent package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Let's load our dataset using Pandas
# If your dataset is not in the same directory as this file, or it has a different extension, this method will not work.
iris_df = pd.read_csv("iris_dataset.csv")
```

```python
# The Pandas method .head() is a nice way to check that your dataset loaded properly
# Does this look similar to your original Excel file?
iris_df.head()
```

```python
# How large is our dataset?
# The first number is the number of rows. The second number is the number of columns.
iris_df.shape
```

```python
# Does our dataset have any empty "cells?"
# If the "non-null count" is different than the shape of your dataset, you have empty cells!
# What does "Dtype" tell us?
iris_df.info()
```

```python
# Pandas .describe() is a great way to get a first glance at the data
# Note that we are missing the column "variety." Why?
iris_df.describe()
```

```python
# How many datapoints do we have of each species?
iris_df["variety"].value_counts()
```

```python
# Let's make a fancy graph! Each point's color is based on its classification (variety)
# Can we see a trend?
sns.FacetGrid(iris_df, hue="variety", height=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
```

```python
# What about petals?
sns.FacetGrid(iris_df, hue="variety", height=5).map(plt.scatter, "petal_length", "petal_width").add_legend()
```

```python
# Let's quickly look for a correlation between variables
sns.heatmap(iris_df.corr())
```

```python
# Let's try another kind of plot
sns.boxplot(x="variety", y="petal_length", data=iris_df)
```

```python
# Let's plot EVERYTHING at once
sns.pairplot(iris_df, hue="variety", height=6)
```
